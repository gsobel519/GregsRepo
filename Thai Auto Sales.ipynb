```python
%%capture
from kls import bdh, bdh_bulk, get_kls_series
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
import datetime as dt
import datetime
from bbg.sapi import BbgSapi
sapi = BbgSapi()
from collections import defaultdict
```

```python
# Seasonally Adjusted Indices

df = bdh_bulk({'vehicle_sales':'THVPRS Index'}, '19900101','20250101')
```

```python
# Plot Overall Level

ax, figure = plt.subplots(figsize = (7,3))

plt.plot(df.vehicle_sales.tail(60));

plt.title('Thai Auto Sales');
```

```python
# Date Ranges

_2016 = df.loc['2016-1-1':'2016-12-31']
_2017 = df.loc['2017-1-1':'2017-12-31']
_2018 = df.loc['2018-1-1':'2018-12-31']
_2019 = df.loc['2019-1-1':'2019-12-31']
_2020 = df.loc['2020-1-1':'2020-12-31']
_2021 = df.loc['2021-1-1':'2021-12-31']
```

```python
ax, figure = plt.subplots(figsize = (7,3))

x_nums = range(0, 12)

x_nums_new = range(len(_2021))

x_names = ['J', 'F', 'M', 'A', 'M', 'J', 'J', 'A', 'S', 'O', 'N', 'D']

plt.xticks(x_nums, x_names);

plt.plot(x_nums, _2016, label = '2016');
plt.plot(x_nums, _2017, label = '2017');
plt.plot(x_nums, _2018, label = '2018');
plt.plot(x_nums, _2019, label = '2019');
plt.plot(x_nums, _2020, label = '2020');
plt.plot(x_nums_new, _2021, label = '2021', marker = '*');


plt.title('Thai Auto Sales');

plt.legend(bbox_to_anchor = (1, 0.77));
```
